```python
import pandas as pd
```

```python
income_df = pd.read_csv('data/median_housing_prices.csv')
housing_df = pd.read_csv('data/median_income.csv')
```

```python
df = pd.merge(income_df, housing_df, on='State', how='inner')
```

```python
df
```

```python
df["affordability"] = df["HouseholdIncome"] / df["MedianValue"]
```

```python
df
```

```python
df = df.sort_values(by=['affordability'])
```

```python
import matplotlib.pyplot as plt
```

```python
f, ax = plt.subplots(figsize=(18,10))
plt.bar(df["State"], df["affordability"])
plt.xticks(rotation=90)
plt.title("State Housing Affordability")
plt.ylabel("Median house price % of median income")
plt.xlabel("State")
plt.savefig("out.png", facecolor="white", edgecolor='none', bbox_inches='tight')
None
```
